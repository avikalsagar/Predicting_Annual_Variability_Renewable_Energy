```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pytz
```

```python
df = pd.read_csv("merged_dataset_SPP.csv")
```

```python
df
```

```python
# Renaming the first column (taken directly from thr GitHub code)
df.rename({'GMT MKT Interval': "DatetimeGMT"}, axis=1, inplace=True)
# Converting this column into Pandas datetime
df['DatetimeGMT'] = pd.to_datetime(df['DatetimeGMT'])
```

```python
# Performing the merge as explained in previous code cell
df['Coal'] = df[' Coal Market'] + df[' Coal Self']
df.drop([' Coal Market', ' Coal Self'], axis=1, inplace=True)

df['Diesel'] = df[' Diesel Fuel Oil Market'] + df[' Diesel Fuel Oil Self']
df.drop([' Diesel Fuel Oil Market', ' Diesel Fuel Oil Self'], axis=1, inplace=True)

df['Hydro'] = df[' Hydro Market'] + df[' Hydro Self']
df.drop([' Hydro Market', ' Hydro Self'], axis=1, inplace=True)

df['Gas'] = df[' Natural Gas Market'] + df[' Gas Self']
df.drop([' Natural Gas Market', ' Gas Self'], axis=1, inplace=True)

df['Nuclear'] = df[' Nuclear Market'] + df[' Nuclear Self']
df.drop([' Nuclear Market', ' Nuclear Self'], axis=1, inplace=True)

df['Solar'] = df[' Solar Market'] + df[' Solar Self']
df.drop([' Solar Market', ' Solar Self'], axis=1, inplace=True)

df['Waste'] = df[' Waste Disposal Services Market'] + df[' Waste Disposal Services Self']
df.drop([' Waste Disposal Services Market', ' Waste Disposal Services Self'], axis=1, inplace=True)

df['Wind'] = df[' Wind Market'] + df[' Wind Self']
df.drop([' Wind Market', ' Wind Self'], axis=1, inplace=True)

df['Waste heat'] = df[' Waste Heat Market'] + df[' Waste Heat Self']
df.drop([' Waste Heat Market', ' Waste Heat Self'], axis=1, inplace=True)

df['Other'] = df[' Other Market'] + df[' Other Self']
df.drop([' Other Market', ' Other Self'], axis=1, inplace=True)
```

```python
df['DatetimeGMT'] = df['DatetimeGMT'].dt.tz_convert(pytz.timezone('US/Eastern'))
df['Date'] = df['DatetimeGMT'].dt.strftime('%Y-%m-%d %H:%M:%S')
df
```

```python
df.info()
```

```python
df.drop('DatetimeGMT', axis = 1, inplace = True)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()
df = df[~df.index.duplicated(keep='first')]
df
```

```python
df['Hour'] = df.index.hour
df
```

```python
hourly_mean_load = df.groupby([df.index.date, 'Hour']).agg({' Load': 'sum', **{col: 'sum' for col in df.columns if (col != ' Load' and col != 'Hour')}})
hourly_mean_load
```

```python
indices = hourly_mean_load.index
indices
```

```python
final_data = pd.DataFrame(hourly_mean_load, index = indices)
final_data = final_data.reset_index()
final_data['Date'] = pd.to_datetime(final_data['level_0'])
final_data
```

```python
final_data.to_csv("Avg SPP (2018-22).csv")
```
